This project requires you to **design, implement, and critically analyze** a small NLP research study integrating:

This is not a coding exercise. It is a **research-oriented project** requiring:

# **1. Research Question – 2 pts**

### Requirements (1 pts)

* Maximum 150 words
* Must be precise, testable, and measurable
* Must involve at least **one linguistic dimension**

Allowed domains:

* Text classification
* Sequence labeling (POS, NER, chunking)
* Structured prediction
* Language modeling
* Transformer-based analysis

### You must:

  * the task
  * the target variable

### Required Reflection (1 pts)

1. Why is your research question (non-trivial)?
2. What linguistic phenomenon does it involve?
